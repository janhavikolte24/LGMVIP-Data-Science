```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("image.jpg")
plt.imshow(image)
cv2.waitKey(0)
```

```python
# now converting to grey scale 
gr_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gr_img)
cv2.waitKey(0)
```

```python
# now converting it into an inverted image 
inv_img = 255 - gr_img
plt.imshow(inv_img)
cv2.waitKey(0)
```

```python
# now blur the image 
blurr_img = cv2.GaussianBlur(inv_img,(21,21),0)
```

```python
# invert the blurred image 
inv_blur = 255 - blurr_img
pencil_sketch = cv2.divide(gr_img,inv_img,scale= 450.0)
plt.imshow(pencil_sketch)
cv2.waitKey(0)
```
